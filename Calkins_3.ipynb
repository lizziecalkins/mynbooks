# Exercise 3-1: Create some plots

```python
import pandas as pd
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

```python
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

## Visualize the data

```python
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate')
```

```python
mortality_wide.plot()
```

```python
mortality_wide.plot.line(y='15-19 Years', title='Child Death Rates:15-19', legend=False, grid=True, rot=45)
```

```python
mortality_wide.plot.area(legend='reverse', grid=True)
```

```python
mortality_wide.query('Year == 2000').plot.bar(title='Child Death Rates in the Year 2000', xlabel='',rot=70)
```

```python
mortality_data.query('Year == 2000').pivot(index='Year', columns='AgeGroup', values='DeathRate').plot.bar(title='Child Death Rates in the Year 2000', xlabel='',rot=70)
```

```python
mortality_data.plot.hist(y='DeathRate', bins=15, legend= True, rot=70)
```

```python
mortality_data.query('Year == 1900').plot.density(y='DeathRate',title='Death Rates in the Year 1900')
```

```python
mortality_data.query('Year in [1900, 1925, 1950, 1975, 2000]').groupby('Year').DeathRate.sum().plot.pie(legend=True)
```

```python
mortality_wide.query('Year in [1900, 1925, 1950, 1975, 2000]').plot.barh(title=
['Child Death Rates for Ages 01-04',
 'Child Death Rates for Ages 05-09',
 'Child Death Rates for Ages 10-14',
 'Child Death Rates for Ages 15-19'],sharey=True, grid=True, legend=True, subplots=True, layout=(2,2), figsize=(10,10))
```
