## Other techniques for enhancing a plot

```python
g = sns.relplot(
    data=mortality_data, 
    kind='line', x='Year', y='DeathRate', hue='AgeGroup', aspect=1.25)
for ax in g.axes.flat:
    ax.set_title('Deaths by Age Group (2000-2016)')
    ax.set_xlabel('')
    ax.set_xticks([x for x in range(2000,2017,2)])
    ax.set_xticklabels(['Year ' + str(x) for x in range(0,17,2)])
    ax.tick_params('x', labelrotation=30)
    ax.set_ylabel('Deaths per 100,000')
    ax.set_xlim(2000,2016)
    ax.set_ylim(0,80)
```

```python
ax = sns.lineplot(data=mortality_data, 
                  x='Year', y='DeathRate', hue='AgeGroup')
ax.set_title('Deaths by Age Group')
ax.set_ylabel('Deaths per 100,000')
ax.annotate(text='Spanish Flu Pandemic', 
    xy=(1918, 1650), xytext=(1925, 1900), 
    arrowprops=dict(facecolor='red', width=3, headwidth=12, headlength=6))
```

```python
colors = ['black','red','orange','blue']
sns.set_palette(sns.color_palette(colors))
sns.relplot(data=mortality_data, kind='line',
            x='Year', y='DeathRate', hue='AgeGroup')
```

```python
import numpy as np
import matplotlib.pyplot as plt

death_rates = {
    1900: [30, 20, 50, 150],
    1950: [15, 10, 25, 100],
    2000: [5, 5, 10, 50]}
age_groups = ['0-14', '15-29', '30-44', '45+']
years = list(death_rates.keys())
average_death_rates = np.array([np.mean([death_rates[year][i] for year in years]) for i in range(len(age_groups))])
plt.figure(figsize=(10, 6))
bar_width = 0.25
x = np.arange(len(age_groups))
plt.bar(x - bar_width, average_death_rates, width=bar_width, label='Average Death Rate', color='b')
for i, year in enumerate(years):
    plt.bar(x + (i * bar_width), death_rates[year], width=bar_width, label=str(year))
plt.title('Death Rates by Age Group for the Years 1900, 1950, and 2000')
plt.xlabel('Age Groups')
plt.ylabel('Death Rate')
plt.xticks(x + bar_width, age_groups)
plt.legend()
plt.tight_layout()

plt.gcf().set_size_inches(10, 10 / 1.8)
```

```python
import numpy as np
import matplotlib.pyplot as plt

death_rates = {
    1900: [30, 20, 50, 150],
    1950: [15, 10, 25, 100],
    2000: [5, 5, 10, 50]}
age_groups = ['0-14', '15-29', '30-44', '45+']
years = list(death_rates.keys())
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
for i, year in enumerate(years):
    axes[i].bar(age_groups, death_rates[year], color='b')
    axes[i].set_title(f'Death Rates in {year}')
    axes[i].set_xlabel('Age Groups')
    axes[i].set_ylabel('Death Rate')
    axes[i].set_ylim(0, max(max(death_rates[year]) for year in years) + 20)
plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt

death_rates = {
    1900: [30, 20, 50, 150],
    1950: [15, 10, 25, 100],
    2000: [5, 5, 10, 50]}
age_groups = ['0-14', '15-29', '30-44', '45+']
years = list(death_rates.keys())
years_filtered = [year for year in years if year >= 1950]
death_rates_filtered = {year: death_rates[year] for year in years_filtered}
death_rates_by_age_group = np.array([death_rates_filtered[year] for year in years_filtered])
plt.figure(figsize=(10, 6))
for i, age_group in enumerate(age_groups):
    plt.plot(years_filtered, death_rates_by_age_group[:, i], marker='o', label=age_group)
plt.title('Death Rates by Age Group (1950 to 2000)')
plt.xlabel('Year')
plt.ylabel('Death Rate')
plt.xticks(years_filtered)
plt.legend(title='Age Groups')
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

death_rates = {
    1900: [30, 20, 50, 150],
    1925: [25, 15, 40, 120],
    1950: [15, 10, 25, 100],
    1975: [10, 5, 15, 80],
    2000: [5, 5, 10, 50]}
age_groups = ['0-14', '15-29', '30-44', '45+']
years = list(death_rates.keys())
fig, axes = plt.subplots(3, 2, figsize=(12, 15))
axes = axes.flatten()
for i, year in enumerate(years):
    axes[i].bar(age_groups, death_rates[year], color='b')
    axes[i].set_title(f'Death Rates in {year}')
    axes[i].set_ylabel('Deaths per 100,000')
    axes[i].set_ylim(0, max(max(death_rates[year]) for year in years) + 20)
for j in range(i + 1, len(axes)):
    axes[j].axis('off')
plt.suptitle('Death Rates by Age Group Over Selected Years', fontsize=16, y=1.02)
plt.tight_layout(rect=[0, 0, 1, 0.95]) 
plt.savefig('barCharts.png')
```
